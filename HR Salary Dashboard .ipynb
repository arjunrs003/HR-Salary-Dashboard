Now, we are ready to start modelling. 

```python
def IQR_values(dataframe,column_name):
    q1=dataframe[column_name].quantile(0.25)
    q2=dataframe[column_name].quantile(0.50)
    q3=dataframe[column_name].quantile(0.75)
    IQR=q3-q1 
    lower_limit=q1-1.5*IQR
    upper_limit=q3+1.5*IQR
    return lower_limit,upper_limit


outlier={}
for i in numerical_features.columns:
    lower_limit,upper_limit=IQR_values(numerical_features,i)
    outlier[i]=[lower_limit,upper_limit]
```

```python
outlier
```

```python
continous_features
```

```python
Salary_outliers=(df['Salary']<outlier['Salary'][0])|(df['Salary']>outlier['Salary'][1])
salary_outlier=df.loc[Salary_outliers].index
salary_outlier
```

```python
df.drop(salary_outlier,inplace=True)
```

```python
#we need to convert the date time into numerical value.
import datetime as dt
df['DOB']=df['DOB'].map(dt.datetime.toordinal)
df['DateofHire']=df['DateofHire'].map(dt.datetime.toordinal)
df['DateofTermination']=df['DateofTermination'].map(dt.datetime.toordinal)
df['LastPerformanceReview_Date']=df['LastPerformanceReview_Date'].map(dt.datetime.toordinal)
```

```python
df.head(2)
```

```python
#Let us standardise our data
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
x=df.drop(['Salary','Employee_Name'],axis=1)
x
```

```python
y=df['Salary']
y
```

```python
x=pd.DataFrame(std.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
#let us import ML libraries
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
import sklearn.metrics as metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Model Building
algos = [LinearRegression(),  Ridge(), Lasso(),
          KNeighborsRegressor(), DecisionTreeRegressor()]

names = ['Linear Regression', 'Ridge Regression', 'Lasso Regression',
         'K Neighbors Regressor', 'Decision Tree Regressor']

rmse_list = []
R2=[]
```

```python
for name in algos:
    model = name
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    MSE= metrics.mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(MSE)
    rmse_list.append(rmse)
    r2 = r2_score(y_test,y_pred)
    R2.append(r2)
    print(model,y_pred)
   
```

```python
evaluation = pd.DataFrame({'Model': names,
                           'RMSE': rmse_list,
                           'R_Square': R2})
```

```python
evaluation
```

We can see that Decision Tree has lowest RMSE score, we can confirm Decision Tree has our final model and better r square value than others

# Let us do hyper parameter tunning on linear regression

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
LinearRegression().get_params()
```

```python
param_dist = {'copy_X': [True,False],
              'fit_intercept': [True,False],
              'n_jobs': [5,10,20,30,50,100] ,
              'normalize': [True,False],
              'positive': [True,False]}
```

```python
LinearRegression_cv = RandomizedSearchCV(LinearRegression(), param_dist, cv = 5)
```

```python
LinearRegression_cv.fit(x_train,y_train)
```

```python
print("Tuned Decision Tree Parameters: {}".format(LinearRegression_cv.best_params_))
print("Best score is {}".format(LinearRegression_cv.best_score_))
```
